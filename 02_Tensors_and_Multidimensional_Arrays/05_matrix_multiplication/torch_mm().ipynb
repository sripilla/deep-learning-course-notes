```python

"""
torch.mm() performs matrix multiplication between two tensors.
(m × n) × (n × p) → (m × p)
torch.mm(Tensor_1, Tensor_2, out=None)
works with 2D Matrices.
Doesn't support broadcasting (). Smaller tensor is expanded to match larger one
but mm() doesnt support this
"""
import torch

mat_1 = torch.tensor([
   [1, 2, 3],
   [4, 3, 8],
   [1, 7, 2]
])

mat_2 = torch.tensor([
   [2, 4, 1],
   [1, 3, 6],
   [2, 6, 5]
])

torch.mm(mat_1, mat_2)
```

```python
mat_1 = torch.tensor([
   [1, 2],
   [4, 3]
])   # (2×2)

mat_2 = torch.tensor([
   [2, 4, 1],
   [1, 3, 6]
])   # (2×3)

torch.mm(mat_1, mat_2)
```
