```python
import torch

tensor_A = torch.rand(size=(2,3))
tensor_B = torch.rand(size=(2,3))

# This will FAIL, Matrix multiplication â‰  Element-wise multiplication
# torch.matmul(tensor_A, tensor_B)

# Fix using transpose
tensor_C = torch.matmul(tensor_A, tensor_B.T)
print(tensor_C)

"""
Use: torch.matmul() or @
Rule:
(đť‘šĂ—đť‘›)(đť‘›Ă—đť‘ť)=(đť‘šĂ—đť‘ť)
(mĂ—n)(nĂ—p)=(mĂ—p)
Condition:
đź‘‰ cols(A) == rows(B)
If shapes mismatch:
đź‘‰ Use transpose â†’ B.T
Example:
(2Ă—3) Ă— (2Ă—3) âťŚ â†’ (2Ă—3) Ă— (3Ă—2) âś… â†’ (2Ă—2)
Important:
đź‘‰ * = element-wise âťŚ
đź‘‰ matmul / @ = matrix multiplication âś…
Methods:
mm() â†’ only 2D
matmul() â†’ all cases
bmm() â†’ batch
@ â†’ clean syntax
Core idea:
đź‘‰ Rows Ă— Columns â†’ dot product
"""
```

```python
import torch

# -------------------------------
# đź”ą 1. Vector Ă— Vector (Dot Product)
# -------------------------------
a = torch.tensor([1., 2., 3.])
b = torch.tensor([4., 5., 6.])

# Result: scalar (dot product)
print("Vector Ă— Vector:", torch.matmul(a, b))


# -------------------------------
# đź”ą 2. Matrix Ă— Vector
# -------------------------------
A = torch.tensor([[1., 2., 3.],
                  [4., 5., 6.]])

v = torch.tensor([1., 2., 3.])

# Result: vector
print("Matrix Ă— Vector:\n", torch.matmul(A, v))


# -------------------------------
# đź”ą 3. Matrix Ă— Matrix
# -------------------------------
A = torch.tensor([[1., 2.],
                  [3., 4.]])

B = torch.tensor([[5., 6.],
                  [7., 8.]])

# Result: matrix
print("Matrix Ă— Matrix:\n", torch.matmul(A, B))


# -------------------------------
# đź”ą 4. Batch Matrix Multiplication (3D tensors)
# -------------------------------
A = torch.rand(2, 2, 3)  # (batch=2, 2x3)
B = torch.rand(2, 3, 4)  # (batch=2, 3x4)

# Result: (2, 2, 4)
print("Batch MatMul shape:", torch.matmul(A, B).shape)


# -------------------------------
# đź”ą 5. Using @ Operator (shortcut)
# -------------------------------
A = torch.tensor([[1., 2.],
                  [3., 4.]])

B = torch.tensor([[5., 6.],
                  [7., 8.]])

print("Using @ operator:\n", A @ B)
```
