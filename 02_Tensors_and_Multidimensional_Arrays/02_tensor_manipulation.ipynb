```python

import torch
import numpy as np

#  Input Data
data1 = [1, 2, 3, 4, 5, 6]
data2 = np.array([1.5, 3.4, 6.8, 9.3, 7.0, 2.8])

#  Tensor Creation
t1 = torch.tensor(data1)
t2 = torch.Tensor(data1)
t3 = torch.as_tensor(data2)
t4 = torch.from_numpy(data2)

#  Printing Results
print("Tensor:", t1, "Data type:", t1.dtype)
print("Tensor:", t2, "Data type:", t2.dtype)
print("Tensor:", t3, "Data type:", t3.dtype)
print("Tensor:", t4, "Data type:", t4.dtype)
print("numpy array", data2, "Data type:", data2.dtype)



```

```python
import numpy as np          # Import NumPy for array operations
import torch               # Import PyTorch for tensor operations

# Create a NumPy array (2x2 matrix)
npx = np.array([[1, 2], [3, 4]])

# Convert NumPy array → PyTorch tensor
# NOTE: This shares memory (no copying happens)
ty = torch.from_numpy(npx)

# Print original NumPy array
print("npx=", npx)

# Print PyTorch tensor created from NumPy array
print("ty=", ty)

# Print data type of NumPy array (default is int64)
print("npx.dtype=", npx.dtype)

# Print data type of tensor (matches NumPy dtype → torch.int64)
print("ty.dtype=", ty.dtype)

# Convert PyTorch tensor → NumPy array
# Again, this shares memory (no copy)
npz = ty.numpy()

# Print converted NumPy array
print("npz=", npz)
```
